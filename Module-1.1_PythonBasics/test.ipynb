Operation in python

```python
x = 10; y = 3
sum = x + y # summation 
sub = x - y # subtraction
mult = x * y # multiplication
div = round(x/y, 2)
# div = round(x / y , 2) # division 

rem = x % y # remainder 

print(sum , sub , mult, div , rem)
```

```python
x=10; y=3

greater_than = x>y
greater_than_equal = (10 >= 10)
print(greater_than, greater_than_equal)


less_than = x < y
less_than_equal = (10 <= 10)
print(less_than, less_than_equal)

equal = x==y
print(equal)


not_equal = x!=y
print(not_equal)
```

Probelm: Digit Summation

```python
inp = input()
numbers = inp.split()

x = int(numbers[0])
y = int(numbers[1])

last_digit_of_x = x % 10
last_digit_of_y = y % 10

sum = last_digit_of_x + last_digit_of_y

print(sum)
```
